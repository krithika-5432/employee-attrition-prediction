```python
# Load Dataset

df = pd.read_csv("processed-data.csv")
df.head()
```

```python
# Convert Target Variable

df["Attrition"] = df["Attrition"].map({"Yes":1, "No":0})
df["Attrition"].value_counts()
```

```python
# Income per Experience Year

df["IncomePerYear"] = df["MonthlyIncome"] / (df["TotalWorkingYears"] + 1)
```

```python
# Promotion Gap

df["PromotionGap"] = df["YearsAtCompany"] - df["YearsSinceLastPromotion"]
```

```python
# Company Experience Ratio

df["CompanyExperienceRatio"] = df["YearsAtCompany"] / (df["TotalWorkingYears"] + 1)
```

```python
# Manager Stability

df["ManagerStability"] = df["YearsWithCurrManager"] / (df["YearsAtCompany"] + 1)
```

```python
# Workload Indicator

df["WorkloadScore"] = df["JobInvolvement"] + df["WorkLifeBalance"]
```

```python
# Salary Growth Indicator

df["SalaryGrowth"] = df["PercentSalaryHike"] / 100
```

```python
# Identify Categorical Columns

categorical_cols = df.select_dtypes(include=["object"]).columns
categorical_cols
```

```python
# Encode Categorical Variables

categorical_cols = df.select_dtypes(include=["object"]).columns
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
```

```python
# Feature Scaling

from sklearn.preprocessing import StandardScaler
X = df.drop("Attrition", axis=1)
y = df["Attrition"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

df_scaled = pd.DataFrame(X_scaled, columns=X.columns, index=df.index)
df_scaled["Attrition"] = y

df = df_scaled
```
