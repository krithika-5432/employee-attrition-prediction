```python
# Load the dataset

import pandas as pd
df = pd.read_csv("feature-engineered-data.csv")
df.head()
```

```python
# Create feature_columns.pkl

import joblib
feature_columns = df.drop("Attrition", axis=1).columns.tolist()
joblib.dump(feature_columns, "feature_columns.pkl")
print("feature_columns.pkl saved")
```

```python
# Create scaler.pkl

from sklearn.preprocessing import StandardScaler
X = df.drop("Attrition", axis=1)
scaler = StandardScaler()
scaler.fit(X)
joblib.dump(scaler, "scaler.pkl")
print("scaler.pkl saved")
```

```python
# Create shap_explainer.pkl

import shap
import joblib
model = joblib.load("xgboost.pkl")
X = df.drop("Attrition", axis=1)
explainer = shap.TreeExplainer(model)
joblib.dump(explainer, "shap_explainer.pkl")
print("shap_explainer.pkl saved")
```
